##### Counting the occurance of each items that is in grouped column

```python
# here we will look for any NaN values in grouped column
filtered_df.isna()
```

```python
filtered_df = filtered_df.dropna(how='all')
```

```python
filtered_df.isna()
```

```python
print(filtered_df.dtypes)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in filtered_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
print(count)

# (row_list, 2): this means we are looking for the combinations where 2 items are sold together
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in filtered_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

### Findings: Thus, the group of iPhone, Lightening Charging Cable is mostly sold together. Thus, there is need to have an adequate stock of such products.
